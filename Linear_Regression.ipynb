"The market historical data set of real estate valuation are collected from Sindian Dist., New Taipei City, Taiwan. The â€œreal estate valuationâ€ is a regression problem. The data set was randomly split into the training data set (2/3 samples) and the testing data set (1/3 samples)."

This data set contains the transaction data of the purchase of a house, the house's age in years, the distance from the house to the nearest MRT (mass rapid transit) station in meters, and the house price of unit area."

```python
import pandas as pd
data = pd.read_csv("https://raw.githubusercontent.com/YeshayaG/MCON_141/main/Real%20estate.csv")
data.head()


```

```python
data = data.drop(["No","X5 latitude","X6 longitude"],axis = 1)
data
```

Removed data that I deemed not important.

```python
data.isnull().sum()
```

We see there are no null values in our data set

```python
data.dtypes
```

```python
#data["year"] = data['X1 transaction date'].apply(str).split('.')
data["year"] = data['X1 transaction date'].apply(str).str.split('.').str[0]
data["month"] = data['X1 transaction date'].apply(str).str.split('.').str[-1]
data
```

```python
def real_month(x):
  if x == "0":
    return "1"
  elif x == "083":
    return "2"
  elif x == "167":
    return "3"
  elif x == "25":
    return "4"
  elif x == "333":
    return "5"
  elif x == "417":
    return "6"
  elif x == "5":
    return "7"
  elif x == "583":
    return "8"
  elif x == "667":
    return "9"
  elif x == "75":
    return "10"
  elif x == "833":
    return "11"
  elif x == "917":
    return "12"
  else:
    return None
data["month"] = data["month"].apply(real_month)
```

The data is initially in a werid format; therefore, I converted it into a year column and a month column.

```python
data = data.drop(["X1 transaction date"],axis = 1)
data = pd.get_dummies(data,drop_first=True)
data
```

I converted the year and month column to one hot encoding in order to use in our linear regression. Note that that only years 2012 and 2013 are in the data set, so 2012 = 0 and 2013 =1.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

train, test = train_test_split(data,test_size=.2)
x = train.drop(["Y house price of unit area"],axis = 1)
y = train["Y house price of unit area"]
model = LinearRegression().fit(X=x,y=y)


```

We split the data to both a train and a test set in order to evaluate the regression model on data that wasn't used to fit the model.

```python
x = test.drop(["Y house price of unit area"],axis = 1)
y = test["Y house price of unit area"]
model.score(x,y)
```

The score we got is above. We see that our x's are somewhat related to our y value.
