"This dataset contains current estimates (live population clock), historical data, and projected figures of world countries and dependent territories. Data based on the latest United Nations Population Division estimates."

```python
import pandas as pd
import numpy as np
data =pd.read_csv("https://raw.githubusercontent.com/YeshayaG/MCON_141/main/world_population.csv")
```

Country/Other - Name of countries and dependent territories.
Population (2020) - Population in the year 2020
Yearly Change - Percentage Yearly Change in Population
Net Change - Net Change in Population
Density (P/Km²)- Population density (population per square km)
Land Area (Km²) - Land area of countries / dependent territories.
Migrants (net) - Total number of migrants
Fert. Rate - Fertility rate
Med. Age - Median age of the population


```python
data.index = np.arange(1,len(data)+1)
data.drop(["Urban Pop %","World Share"],axis=1,inplace=True)
data
```

```python
data.isna().sum()
```

```python
data.dtypes
```

```python
data.describe()
```

As expected China and India have the greatest populations. Note that the populations are expressed in scientific notation. 

```python
import matplotlib.pyplot as plt
data.plot.bar(x="Country/Other",y="Population (2020)",figsize=(40,10))
plt.hlines(38776610, xmin=0, xmax=500000 )
```

We see from this scatter plot that most countries have a very dense population.

```python
import matplotlib.pyplot as plt
data.plot.scatter(x='Population (2020)',y ='Land Area (Km²)')
plt.scatter(data['Population (2020)'], data['Land Area (Km²)'],s=20)
```

This is a zoomed in scatter plot.

```python
plt.scatter(x = data['Population (2020)'],y = data['Land Area (Km²)'],s=20)
plt.xlim(0,18861980)
plt.ylim(0,216400)
```
